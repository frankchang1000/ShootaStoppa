```python
import os
```

```python
training_image_files = os.listdir("data/Images")
```

```python
for image_file in training_image_files:
    file_type = os.path.split(image_file)[-1]
    file_name, file_type = os.path.splitext(file_type)
    # change the file name to .jpg
    if file_type != ".jpg":
        os.rename("data/Images/" + image_file, "data/Images/" + file_name + ".jpg")
        print("Renamed " + image_file + " to " + image_file[:-3] + "jpg")
```

```python
images_files = os.listdir("data/Images")
labels_files = os.listdir("data/Labels")

image_base_path = "data/Images/"
label_base_path = "data/Labels/"

for i, image_file in enumerate(images_files):
    old_file_name = os.path.split(image_file)[-1]
    file_name, file_type = os.path.splitext(old_file_name)

    print(file_type)
    print(file_name)

    new_file_name = str(i).zfill(6)

    print(new_file_name)

    old_image_file = old_file_name
    old_label_file = file_name + ".xml"

    new_image_file = new_file_name + ".jpg"
    new_label_file = new_file_name + ".xml"

    os.rename(image_base_path + old_image_file, image_base_path + new_image_file)
    os.rename(label_base_path + old_label_file, label_base_path + new_label_file)
```

```python
training_image_files = os.listdir("data/Images")

with open("data/train.txt", "w") as f:
    for image_file in training_image_files:
        f.write(image_file[0:-4] + "\n")
        print("Wrote " + image_file + " to train.txt")
```
